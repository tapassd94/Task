```python
import pandas as pd
pd.set_option('display.max_columns',100)
from flaml import AutoML
from sklearn.model_selection import train_test_split
```

```python
train=pd.read_csv('/DATA/HOME_DIR/ramg/Ram/other/train_BRCpofr.csv')
train
```

```python
qualification = pd.get_dummies(train['qualification'])
area = pd.get_dummies(train['area'])
income	 = pd.get_dummies(train['income'])
num_policies = pd.get_dummies(train['num_policies'])
policy = pd.get_dummies(train['policy'])
type_of_policy = pd.get_dummies(train['type_of_policy'])
train = train.drop(['qualification','income','num_policies','policy','type_of_policy','area','id'],axis = 1)
train = train.join([area,qualification,income,num_policies,policy,type_of_policy])
train['gender']=train['gender'].apply(lambda x: 1 if x=='Male' else 0)
train  
```

```python
test=pd.read_csv('/DATA/HOME_DIR/ramg/Ram/other/test_koRSKBP.csv')
test
```

```python
qualification = pd.get_dummies(test['qualification'])
area = pd.get_dummies(test['area'])
income	 = pd.get_dummies(test['income'])
num_policies = pd.get_dummies(test['num_policies'])
policy = pd.get_dummies(test['policy'])
type_of_policy = pd.get_dummies(test['type_of_policy'])
test = test.drop(['qualification','income','num_policies','policy','type_of_policy','area','id'],axis = 1)
test = test.join([area,qualification,income,num_policies,policy,type_of_policy])
test['gender']=test['gender'].apply(lambda x: 1 if x=='Male' else 0)
test  
```

```python
train=train.rename({'2L-5L':'2_5','5L-10L':'5_10','<=2L':'2'},axis=1)
test=test.rename({'2L-5L':'2_5','5L-10L':'5_10','<=2L':'2'},axis=1)
test
```

```python
y=train['cltv']
x=train.drop('cltv',axis=1)
```

```python
automl = AutoML()
settings = {
    "time_budget": 1200,  # total running time in seconds
    "metric": 'r2',  # primary metrics for regression can be chosen from: ['mae','mse','r2','rmse','mape']
    "task": 'regression',  
    #"estimator_list":["lgbm"],    
    "log_file_name": 'log.log',  # flaml log file
}
```

```python
automl.fit(X_train=x, y_train=y, **settings)
```

```python
automl.best_estimator
```

```python
fi=pd.concat([pd.DataFrame(test.columns),pd.DataFrame(automl.model.estimator.feature_importances_)],axis=1)
fi.columns=['feature','importance']
fi
```

```python
pd.concat([pd.read_csv('/DATA/HOME_DIR/ramg/Ram/other/test_koRSKBP.csv',usecols=['id']),pd.DataFrame(automl.predict(test))],axis=1).to_csv('sub.csv')
```
